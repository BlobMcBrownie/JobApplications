```python
import csv
import random
from random import choice
```

```python
####### UNIVERSITY #########

# uniCode (0, 9362)

inputName = "world-unis.csv"
outputName = "university_table.sql"
insertHeader = "INSERT INTO University (country,uniName,website,uniCode,QSranking)"
insertHeader += "\nVALUES\n"

longestUniName = 0

# single quotes BETTER
def transform_row(row):
    global longestUniName
    ret = ""
    ret += f"'{row[0]}',"
    row[1] = row[1].replace("'", "''")
    ret += f"'{row[1]}'," # 114
    ret += f"'{row[2]}'," # 66
    ret += f'{row[3]},'
    ret += f'{row[4]}'
    longestUniName = max(longestUniName, len(row[1]))
    return ret

with open(inputName, 'r', encoding = "utf8") as infile:
    reader = csv.reader(infile)
    nrows = sum(1 for row in reader)
    print(f"{nrows = }")

with open(inputName, 'r', encoding = "utf8") as infile:
    with open(outputName, 'w', encoding = "utf8") as outfile:
        reader = csv.reader(infile)
        outfile.write(insertHeader)
        for i, row in enumerate(reader):
            if i == nrows - 1:
                outfile.write("\t(" + transform_row(row) + ");" + "\n")
            else:
                outfile.write("\t(" + transform_row(row) + ")," + "\n")

print(f"{longestUniName = }")
```

```python
####### COOPERATION #########

inputName = "world-unis.csv"
outputName = "cooperation_table.sql"
insertHeader = "INSERT INTO Cooperation (uniCode1,uniCode2)"
insertHeader += "\nVALUES\n"

maxCode = 9362
numberOfRowsWanted = 100

def getRandomNot(first):
    second = random.randint(0, maxCode)
    while first == second:
        second = random.randint(0, maxCode)
    return second

with open(inputName, 'r', encoding = "utf8") as infile:
    with open(outputName, 'w', encoding = "utf8") as outfile:
        reader = csv.reader(infile)
        outfile.write(insertHeader)
        for i in range(numberOfRowsWanted):
            if i == numberOfRowsWanted - 1:
                outfile.write(f"\t({i}, {getRandomNot(i)});\n")
            else:
                outfile.write(f"\t({i}, {getRandomNot(i)}),\n")
```
